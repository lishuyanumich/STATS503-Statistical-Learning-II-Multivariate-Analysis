(a) We are given n = 7 observations in p = 2 dimensions. For each observation, there is an associated class label. Sketch the observations.

```r
# library
library(ggplot2)
```

```r
X1 = c(3,2,4,1,2,4,4)
X2 = c(4,2,4,4,1,3,1)
X = matrix(c(X1,X2), ncol=2)
Y=c("Red", "Red", "Red", "Red", "Blue", "Blue", "Blue")
#In order for the svm() function to perform classification, we must encode the response as a factor variable.
toy.data = data.frame(x = X, y = as.factor(Y))
toy.data        
```

```r
# Sketch the observations.
colors = Y
plot(X1, X2, col = colors, xlim = c(0, 5), ylim = c(0, 5))
```

(b) Sketch the optimal separating hyperplane, and provide the equation for this hyperplane (of the form (9.1)).

```r
library(e1071)
svmfit=svm(y~x.2+x.1, data=toy.data, kernel="linear", cost=10, scale=FALSE)
plot(svmfit,toy.data)
```

Note that the two arguments to the `plot.svm()` function are the output of the call to `svm()`, as well as the data used in the call to svm(). The region of feature space that will be assigned to the Blue class is shown in light cream colour, and the region that will be assigned to the Red class is shown in darkish red. The decision boundary between the two classes is linear (because we used the argument `kernel=linear`). We see that in this case no observation is misclassified. We see here that there are 3 support vectors. We can determine their identities as follows:

```r
t(svmfit$coefs) %*% svmfit$SV
```

```r
# intercept
-svmfit$rho
```

So the equation for this hyperplane is 
$$
1-2X_1+2X_2=0
$$

```r
# Sketch the optimal separating hyperplane
plot(X1, X2, col = colors, xlim = c(0, 5), ylim = c(0, 5))
abline(-0.5, 1)
```

(c) Describe the classification rule for the maximal margin classifier. It should be something along the lines of “Classify to Red if $\beta_0 + \beta_1X_1 + \beta_2X_2 > 0$, and classify to Blue otherwise.” Provide the values for $\beta_0 ,\beta_1, \beta_2$.

The optimal separating hyperplane from the previous question has the equation 
$$
1-2X_1+2X_2=0
$$
Visual inspection makes it clear that the prediction for the maximal margin classifier will be Red when $X_2>X_1-0.5$ and be blue when $X_2<X_1-0.5$.

The maximal margin classifier will therefore take the form
$$f(X)=1-2X_1+2X_2$$
where a new observation $X^*$ is assigned to the Red class if $f(X^*)>0$ and the Blue class if $f(X^∗)<0$.

We therefore have 
$$(\beta_0 ,\beta_1, \beta_2)=(1, -2, 1)$$

(d) On your sketch, indicate the margin for the maximal margin hyperplane.

The maximal margin hyperplane is the solid line, and the margin is the distance from the solid line to either of the dashed lines, which is
$$M=\frac{\sqrt{2}}{4}$$

```r
plot(X1, X2, col = colors, xlim = c(0, 5), ylim = c(0, 5))
abline(-0.5, 1)
abline(-1, 1, lty = 2)
abline(0, 1, lty = 2)
```

(e) Indicate the support vectors for the maximal margin classifier.

The support vectors are the points $(2,1)$, $(2,2)$, $(4,3)$ and $(4,4)$.

(f) Argue that a slight movement of the seventh observation would not affect the maximal margin hyperplane.

We can see above that 7th observation is not a support vector. Any small movement of this observation in any direction is not going to change the original optimal separating hyperplane $1-2X_1+2X_2=0$ with the largest margin $M=\frac{\sqrt{2}}{4}$.
