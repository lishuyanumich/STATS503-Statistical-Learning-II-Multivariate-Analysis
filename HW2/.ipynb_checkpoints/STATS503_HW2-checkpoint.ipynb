**2. This problem is about understanding bias and variance in a very simple setting.**

$$\bar{x}=\frac{1}{n}\sum_{i=1}^n x_i$$
$$Bias(\bar{x})=E(\bar{x})-0=E(\frac{1}{n}\sum_{i=1}^n x_i)-0=\frac{1}{n}\sum_{i=1}^nE(x_i)=0$$
$$Var(\bar{x})=Var(\frac{1}{n}\sum_{i=1}^n x_i)=\frac{1}{n^2}\sum_{i=1}^nVar(x_i)=\frac{\sigma^2}{n}$$
$$MSE(\bar{x})=Bias(\bar{x})^2+Var(\bar{x})=\frac{\sigma^2}{n}$$

When $n\rightarrow\infty$, $Bias(\bar{x})=0$, $Var(\bar{x})=\frac{\sigma^2}{n}\rightarrow0$, $MSE(\bar{x})=\frac{\sigma^2}{n}\rightarrow0$.

$$Bias(\bar{x})=E(\bar{x})-0=E(\frac{1}{n}\sum_{i=1}^n x_i)-0=\frac{1}{n}\sum_{i=1}^nE(x_i)=0$$
$$Var(\bar{x})=diag(Cov(\bar{x}))^{1/2}Corr(\bar{x})diag(Cov(\bar{x}))^{1/2}$$
$$Corr(\bar{x})=
\left[ \begin{matrix} 
1 & \rho & \cdots & \rho\\ 
\rho & 1 & \cdots & \rho \\ 
\vdots & \vdots &  \ddots & \vdots \\ 
\rho & \rho & \cdots & 1 \end{matrix} \right]$$

$$
diag(Cov(\bar{x}))^{1/2}=
\left[ \begin{matrix} 
SD(x_1) & 0 & \cdots & 0\\ 
0 & SD(x_2) & \cdots & 0 \\ 
\vdots & \vdots &  \ddots & \vdots \\ 
0 & 0 & \cdots & SD(x_n) \end{matrix} \right]
$$

Thus,
$$
Cov(\bar{x})=
\left[ \begin{matrix} 
Var(x_1) & \rho SD(x_1)SD(x_2) & \cdots & \rho SD(x_1)SD(x_n)\\ 
\rho SD(x_2)SD(x_1) & Var(x_2) & \cdots & \rho SD(x_2)SD(x_n) \\ 
\vdots & \vdots &  \ddots & \vdots \\ 
\rho SD(x_n)SD(x_1) & \rho SD(x_n)SD(x_2) & \cdots & Var(x_n) \end{matrix} \right]\\
=\left[ \begin{matrix} 
\sigma^2 & \rho \sigma^2 & \cdots & \rho \sigma^2\\ 
\rho \sigma^2 & \sigma^2 & \cdots & \rho \sigma^2 \\ 
\vdots & \vdots &  \ddots & \vdots \\ 
\rho \sigma^2 & \rho \sigma^2 & \cdots & \sigma^2 \end{matrix} \right]
=\frac{\sigma^2}{n}+\frac{n(n-1)}{n^2}\rho\sigma^2
$$

Then
$$
MSE(\bar{x})=Bias(\bar{x})^2+Var(\bar{x})
=\frac{\sigma^2}{n}+\frac{n(n-1)}{n^2}\rho\sigma^2
$$
When $n\rightarrow\infty$, $Bias(\bar{x})=0$, $Var(\bar{x})=\frac{\sigma^2}{n}+\frac{n(n-1)}{n^2}\rho\sigma^2\rightarrow \rho\sigma^2$, $MSE(\bar{x})=\frac{\sigma^2}{n}+\frac{n(n-1)}{n^2}\rho\sigma^2\rightarrow \rho\sigma^2$.

0.9n observations follows the original distribution with mean 0 and variance $\sigma^2$. And 0.1n observations follow a distribution with mean $\mu>0$ and variance $\sigma^2$.

$$Bias(\bar{x})=(E(\bar{x})-0)\times0.9+(E(\bar{x})-\mu)\times0.1=0.18\mu$$
$$Var(\bar{x})=Var(\frac{1}{n}\sum_{i=1}^n x_i)=\frac{1}{n^2}\sum_{i=1}^nVar(x_i)=\frac{\sigma^2}{n}$$
$$MSE(\bar{x})=Bias(\bar{x})^2+Var(\bar{x})=(0.1\mu)^2+\frac{\sigma^2}{n}$$

When $n\rightarrow\infty$, $Bias=0.1\mu$, $Var(\bar{x})=\frac{\sigma^2}{n}\rightarrow 0$, $MSE(\bar{x})=(0.1\mu)^2+\frac{\sigma^2}{n}\rightarrow (0.1\mu)^2$.

### KNN Modeling

```r
library(class)
```

We need to pull out the outcome variable and the designing matrix for the training and test data.

```r
train_label = college.train %>% .$Private
train_x = college.train %>% select("Apps","Accept","Enroll","Top10perc","Top25perc","F.Undergrad","P.Undergrad",
                                   "Outstate","Room.Board","Books","Personal","PhD","Terminal","S.F.Ratio",
                                   "perc.alumni","Expend","Grad.Rate")

test_label = college.test %>% .$Private 
test_x = college.test %>% select("Apps","Accept","Enroll","Top10perc","Top25perc","F.Undergrad","P.Undergrad",
                                   "Outstate","Room.Board","Books","Personal","PhD","Terminal","S.F.Ratio",
                                   "perc.alumni","Expend","Grad.Rate")
```

As we can see in the data exploration, the variables have different ranges. It is **very important to scale both training and test data in the same way**. In other words, we need to use same mean and standard deviation to transform both training and test data. Those desired mean and standard deviation used for scaling should only come from the training data.

```r
mean_train = colMeans(train_x)
std_train = sqrt(diag(var(train_x)))
# training data
train_x = scale(train_x, center = mean_train, scale = std_train)
# test data
test_x = scale(test_x, center = mean_train, scale = std_train)
```

- The center = ... argument specifies the mean for scaling.
- The scale = ... arguments specifies the standard deviation used for scaling.
- With the additional arguments, the scale() function will perform
$$
x_{i}^{s t d}=\frac{x_{i}-\text { mean }_{\text {train}}}{\text {std}_{\text {train}}}.
$$

Now the `knn()` function can be used to predict whether `Private` is yes or no.

The function requires four inputs.

- train: matrix or data frame of training set cases.
- test: matrix or data frame of test set cases.
- cl: factor of true classifications of training set, i.e. labels.
- k: number of neighbours considered.

```r
set.seed(42)
pred_train = knn(train_x, 
               train_x, 
               train_label, 
               k = 1)

pred_test = knn(train_x, 
               test_x, 
               train_label, 
               k = 1)
```

The `table()` function can be used to produce a **confusion matrix** in order to determine how many observations were correctly or incorrectly classified.

```r
# Pull out the true responses for the test data
table(pred_train, train_label)
mean(pred_train == train_label)

table(pred_test, test_label)
mean(pred_test == test_label)
```

The testing results using `K = 1` are not very good compared to training errors, which means we have strong overfit. We further seek a best `K`.

```r
k_range = c(5, 10, 20, 25, 50, 200, 400)
train_error = c()
test_error = c()
for(i in 1:length(k_range)){
  pred_train <- knn(train_x, 
               train_x, 
               train_label, 
               k = k_range[i])
  train_error[i] = mean(pred_train != train_label)
  pred_test = knn(train_x, 
               test_x, 
               train_label, 
               k = k_range[i])
  test_error[i] = mean(pred_test != test_label)
}
```

For each iteration in the for loop, we fit a KNN regression model using the same dataset, but with different choice of `K` as k[i]. We compute the training and testing error rate each round.

Let’s create a plot for all those training and test MSEs. Recall we want to plot MSE vs 1/K as smaller K corresponds to more complex models.

```r
errors = data.frame(train_error, test_error, k_range)
ggplot(errors, aes(x = 1/k_range)) + 
  geom_line(aes(y = train_error), col = "red") + geom_point(aes(y = train_error), col = "red") +
  geom_line(aes(y = test_error), col = "blue") + geom_point(aes(y = test_error), col = "blue") +
  ylab("Error Rate") + xlab("1/K") + ggtitle("Training and test error rate for KNN") + theme_minimal()
errors
```

```r
k_range[which(train_error==min(train_error))]
min(train_error)
```

```r
k_range[which(test_error==min(test_error))]
min(test_error)
```

When $k=5$, train error is minimized. When $k=10$, test error is minimized. Smaller the k is, the smaller train error is. So we need to pay more attention to the k which minimizes test_error. And the minimized test error is $0.0641025641025641$.

### LOOCV

```r
# LOOCV using inbuilt knn.cv function 
set.seed(800)
lcv_error = numeric(150)

for(i in 1:150){
  K = knn.cv(college.train[, -1], cl=as.numeric(train_label), k=i)
  lcv_error[i] = mean(as.numeric(train_label)!=K)
}
```

### k-Fold Cross-Validation
####  k=5 and k=10

```r
library(FNN)
# Splitting dataset into training and test data¶
set.seed(1)
train_id = sample(1:nrow(college),floor(0.7*nrow(college)))
train = college[train_id,]
test = college[-train_id,]
```

```r
# K-fold CV on training set
Kfold_CV_knn <- function(K,K_knn,train,train_label){
  fold_size = floor(nrow(train)/K)
  cv_error = rep(0,K)
  for(i in 1:K){
    # iteratively select K-1 folds as training data in CV procedure, remaining as test data.
    if(i!=K){
      CV_test_id = ((i-1)*fold_size+1):(i*fold_size)
    }else{
      CV_test_id = ((i-1)*fold_size+1):nrow(train)
    }
    CV_train = train[-CV_test_id,]
    CV_test = train[CV_test_id,]
    # calculate the mean and standard deviation using CV_train
    mean_CV_train = colMeans(CV_train)
    sd_CV_train = apply(CV_train,2,sd)
    # normalize the CV_train and CV_test using above mean and sd
    CV_train = scale(CV_train,center = mean_CV_train,scale = sd_CV_train)
    CV_test = scale(CV_test,center = mean_CV_train,scale = sd_CV_train)
    # Fit knn
    pred_CV_test = knn(CV_train,CV_test,train_label[-CV_test_id],k = K_knn)
    # Calculate CV error by taking averages
    cv_error[i] = mean(pred_CV_test!=train_label[CV_test_id])
  }
  return(mean(cv_error))
}
```

```r
# Now we set K_fold = 5
set.seed(2020)
K_fold = 5
K_knn = 1:150
cv_error_5 = rep(0,length(K_knn))
for(i in 1:length(K_knn)){
  cv_error_5[i] = Kfold_CV_knn(K = K_fold, K_knn = K_knn[i],train = train[,-1],train_label = train$Private)
}
```

When K=5, which k gives the minimum CV error?

```r
min_5cv_error = min(cv_error_5)
best_k_5 = max(which(cv_error_5 == min(cv_error_5)))
print(paste("The minimal error is", min_5cv_error, ", and the best k is", best_k_5))
```

```r
# Then we set K_fold = 10
set.seed(2020)
K_fold = 10
cv_error_10 = rep(0,length(K_knn))
for(i in 1:length(K_knn)){
  cv_error_10[i] = Kfold_CV_knn(K = K_fold, K_knn = K_knn[i],train = train[,-1],train_label = train$Private)
}
```

When K=10, which k gives the minimum CV error?

```r
min_10cv_error = min(cv_error_10)
best_k_10 = max(which(cv_error_10 == min(cv_error_10)))
print(paste("The minimal error is", min_10cv_error, ", and the best k is", best_k_10))
```

Let's make a plot for k v.s. CV error to view clearly.

```r
plot(cv_error_5~K_knn,type='b',col="blue", main = '5(blue)/10(red)-Fold CV error v.s. choice of k in KNN',xlab = 'K',ylab = 'CV error')
lines(cv_error_10~K_knn,type='b',col="red",xlab = 'K',ylab = 'CV error')
```

```r
best_ks = c(best_k_lcv, best_k_5, best_k_10)
min_test_err = rep(0, 3)
for(i in 1:3){
  pred_test = knn(train=train_x,test=test_x, cl=college.train$Private, k=best_ks[i])
  min_test_err[i] = mean(pred_test!=college.test$Private)
}
min_test_err = min_test_err
min_test_err
```

```r
CV = c("LOOCV", "5_Fold_CV", "10_Fold_CV")
best_k = c(best_k_lcv, best_k_5, best_k_10)
min_train_error = c(best_lcv_error, min_5cv_error, min_10cv_error)   
min_test_error = min_test_err
df_error = data.frame(CV, best_k, min_train_error, min_test_error) 
df_error
```

We can see LOOCV gives the best classification result.

The **similarities** among LOOCV and K-fold cross validation:

(a) We both used "resampling" method. This allows us to obtain some additional knowledge that would not have been obtained by fitting the model just once. Another reason to use such methods is to evaluate the test error over many iterations to better judge the model performance.

(b) From the Error-K plot, we can see the curve trends are almost the same. As K increases, error will goes down rapidly to a lowest point and then goes up slowly.

The **differences** among LOOCV and K-fold cross validation:

(a)For calculating validation MSE, the LOOCV method is more stable when compared with the 'validation set' approach. Yet CV has some randomness in it, whereas the LOOCV does not. With different `seed_value` in `setseed(seed_value)`, if we run the 5/ 10-fold CV, we can see different resulting plots for MSE estimation.

(b)LOOCV can be more computationally expensive than to use K-fold cross validation, particularly if the data size is large and also if the model takes substantial time to complete the learning just once.

(c)Compared to K-Fold cross validation method, the other problem with LOOCV is that it can be subject to high variance or overfitting as we are feeding the model almost all the training data to learn and just a single observation to evaluate.
